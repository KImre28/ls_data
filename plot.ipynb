```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyarrow as pa
import datetime

pd.options.plotting.backend = "matplotlib"
```

```python
df = pd.read_parquet('/content/sample_data/LMKS_NM_1981-2023_hour.parquet.gzip')
```

```python
df.tail()
```

```python
df.describe()
```

```python
ax = df.loc[datetime.datetime(1991, 3,21):datetime.datetime(1991, 3,28)].plot(xlabel = "2022.05.[UT]", ylabel = "Neutron counts [%]", yticks=[50, 60, 70, 80, 90, 100, 110], figsize=(6,3), legend=False, grid=True, x_compat=True)
```

```python
fig1, axes1 = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

subset1 = df.loc[datetime.datetime(1991, 3, 21):datetime.datetime(1991, 3, 28)]
subset1.plot(xlabel="Date", ylabel="Neutron counts", ax=axes1[0], yticks=[50, 60, 70, 80, 90, 100, 110], legend=False, grid=True)

subset2 = df.loc[datetime.datetime(1991, 10, 25):datetime.datetime(1991, 11, 1)]
subset2.plot(xlabel="Date", ylabel="Neutron counts", ax=axes1[1], title='Neutron count rates recorded on Lomnicky Stit', yticks=[50, 60, 70, 80, 90, 100, 110], legend=False, grid=True)

axes1[0].grid(True, which='both', linestyle='-', linewidth=0.5)
axes1[1].grid(True, which='both', linestyle='-', linewidth=0.5)

fig2, axes2 = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

subset3 = df.loc[datetime.datetime(1998, 11, 5):datetime.datetime(1998, 11, 12)]
subset3.plot(xlabel="Date", ylabel="Neutron counts", ax=axes2[0], yticks=[50, 60, 70, 80, 90, 100, 110], legend=False, grid=True)

subset4 = df.loc[datetime.datetime(2000, 7, 13):datetime.datetime(2000, 7, 20)]
subset4.plot(xlabel="Date", ylabel="Neutron counts", ax=axes2[1], yticks=[50, 60, 70, 80, 90, 100, 110], legend=False, grid=True)

axes2[0].grid(True, which='both', linestyle='-', linewidth=0.5)
axes2[1].grid(True, which='both', linestyle='-', linewidth=0.5)

fig3, axes3 = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

subset5 = df.loc[datetime.datetime(2003, 10, 26):datetime.datetime(2003, 11, 2)]
subset5.plot(xlabel="Date", ylabel="Neutron counts", ax=axes3[0], yticks=[50, 60, 70, 80, 90, 100, 110], legend=False, grid=True)

axes3[0].grid(True, which='both', linestyle='-', linewidth=0.5)
```
