```python
# converting Date column data (strings with dates) into Datetime objects
feid_df['Datetime'] = pd.to_datetime(feid_df['Date']) + pd.to_timedelta(feid_df['Time'])
feid_df.set_index('Datetime', inplace = True)
feid_df.drop(['Date', 'Time'], axis=1, inplace=True)

feid_df.to_parquet('FEID_1981-2012.parquet.gzip', compression='gzip')
```

```python
feid_df = pd.read_parquet("/media/filip/DATA/zložky/SPACE::LS/data/FEID_1981-2012.parquet.gzip")
```

```python
# getting rid of all rows before '1981-12-01' and with magnitude smaller than 4
feid_df = feid_df[(feid_df.index >= '1981-12-01 00:00:00') & (feid_df['MagnM'] >= 3)]
print(len(feid_df))


# Calculate the time differences between consecutive rows
time_diff = feid_df.index.to_series().diff().dropna()

# Display the smallest time difference
print("Smallest time difference between consecutive rows:", time_diff.min())


# Convert Timedelta values to seconds and calculate average
total_seconds = sum(td.total_seconds() for td in time_diff)
average_seconds = total_seconds / len(time_diff)

# Calculate the average Timedelta value
average_timedelta = pd.Timedelta(seconds=average_seconds)
# Display the average time difference
print("Average Timedelta:", average_timedelta)
```

```python
nm_hdf = pd.read_parquet("/media/filip/DATA/zložky/SPACE::LS/data/LMKS_NM_1981-2023_hour.parquet.gzip")
omni = pd.read_parquet("/media/filip/DATA/zložky/SPACE::LS/data/f10-dst_1981-2023_hour.parquet.gzip")
mdfh = pd.concat([nm_hdf, omni], axis=1)
```

```python
smdfh = mdfh[:"1990-01-01"].copy()

for step in range(0,8):
    treshold = 0.90 + (step * 0.01)

    found_events_inc = []
    found_events_dec = []
    first = True

    for i, row in smdfh.iterrows():
        #print('index: ', i, 'cols:', row)
        next = mdfh[i + datetime.timedelta(hours=1) : i + datetime.timedelta(hours=24)]

        if(min(next['H_COR']) < (row['H_COR'] *  treshold)):
            if first:
                prev = mdfh[i - datetime.timedelta(hours=24) : i - datetime.timedelta(hours=1)]
                if (max(prev['H_COR']) * (2 - treshold) < row['H_COR']):
                    found_events_inc.append(i)
                else:
                    found_events_dec.append(i)
                first = False
        else:
            first = True

    count_within_range = sum(any((timestamp - pd.Timedelta(hours=24) <= feid_df.index) & (feid_df.index <= timestamp + pd.Timedelta(hours=24)))
                            for timestamp in found_events_dec)

    print(round(treshold, 2), "- From found:", len(found_events_dec), "in FEID:", count_within_range, "- success:", int(count_within_range / len(found_events_dec) * 100))
```

```python
smdfh = mdfh[:"1990-01-01"].copy()

for step in range(0,8):
    treshold = 0.90 + (step * 0.01)

    found_events_inc = []
    found_events_dec = []
    first = True
    sleep = 0

    for i, row in smdfh.iterrows():
        next = mdfh[i + datetime.timedelta(hours=1) : i + datetime.timedelta(hours=24)]

        if not first:
                sleep += 1

        if(min(next['H_COR']) < (row['H_COR'] *  treshold)):

            if sleep == 0:
                prev = mdfh[i - datetime.timedelta(hours=24) : i - datetime.timedelta(hours=1)]

                if (max(prev['H_COR']) * (2 - treshold) < row['H_COR']):
                    found_events_inc.append(i)
                else:
                    found_events_dec.append(i)

            first = False

        if sleep == 23:
            first = True
            sleep = 0

    count_within_range = sum(any((timestamp - pd.Timedelta(hours=24) <= feid_df.index) & (feid_df.index <= timestamp + pd.Timedelta(hours=24)))
                            for timestamp in found_events_dec)

    print(round(treshold, 2), "- From found:", len(found_events_dec), "in FEID:", count_within_range, "- success:", int(count_within_range / len(found_events_dec) * 100))
```
